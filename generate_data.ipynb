```python
import pandas as pd
import random
import json

# Sample data
sample_patient = {
    "Name": "Sarah",
    "Age": 42,
    "Gender": "Female",
    "Ethnicity": "Caucasian",
    "Previous treatments": "None",
    "Medical history": "No significant issues",
    "Allergies": "None",
    "Skin type": "Combination, prone to occasional breakouts",
    "Sun exposure history": "Moderate",
    "Smoking status": "Non-smoker",
    "Primary concerns": "Forehead lines, crow's feet, cheek volume loss",
    "Secondary concerns": "Facial asymmetry (left side droopier)",
    "Desired outcome": "Natural-looking rejuvenation",
    "Budget constraints": "Willing to invest in a comprehensive treatment plan",
    "Occupation": "Corporate executive (frequent presentations)",
    "Facial expressions": "Tends to furrow brow when concentrating",
    "Sleep habits": "Sleeps on left side",
    "Recent weight changes": "Lost 10 pounds in the last year",
    "Hormonal status": "Perimenopausal",
    "Physical Examination": {
        "3D facial scan": "Performed",
        "Skin elasticity": "Moderate loss of elasticity",
        "Facial muscle strength": "Slightly weaker on left side",
        "Bone structure": "Slight asymmetry in cheekbone prominence"
    }
}

# Function to generate random patient data
def generate_fake_patient_data(num_patients=100):
    patient_data = []
    for i in range(num_patients):
        # Generate random values based on sample data
        age = random.randint(18, 75)
        gender = random.choice(["Male", "Female"])
        ethnicity = random.choice(["Caucasian", "African American", "Asian", "Hispanic", "Other"])
        treatments = random.choice(["None", "Botox", "Filler", "Chemical Peel", "Laser Treatment"])
        medical_history = random.choice(["No significant issues", "Diabetes", "Hypertension", "Asthma", "Allergic to penicillin"])
        allergies = random.choice(["None", "Penicillin", "Latex", "Aspirin", "Pollen"])
        skin_type = random.choice(["Oily", "Dry", "Combination", "Sensitive"])
        sun_exposure = random.choice(["Low", "Moderate", "High"])
        smoker = random.choice(["Non-smoker", "Smoker"])
        primary_concern = random.choice(["Wrinkles", "Hyperpigmentation", "Sagging skin", "Volume loss"])
        secondary_concern = random.choice(["Facial asymmetry", "Dry skin", "Acne scars", "Dark circles"])
        outcome = random.choice(["Natural-looking rejuvenation", "Smooth skin", "Lifted appearance", "Even skin tone"])
        budget = random.choice(["Minimal", "Willing to invest", "Unlimited"])
        occupation = random.choice(["Teacher", "Engineer", "Corporate executive", "Nurse", "Retired"])
        facial_expressions = random.choice(["Furrows brow", "Smiles often", "Neutral", "Raises eyebrows frequently"])
        sleep_habits = random.choice(["Back sleeper", "Side sleeper (left)", "Side sleeper (right)", "Stomach sleeper"])
        weight_changes = random.choice(["Lost weight", "Gained weight", "No change"])
        hormonal_status = random.choice(["Menopausal", "Perimenopausal", "Normal", "On hormonal therapy"])
        skin_elasticity = random.choice(["High elasticity", "Moderate loss", "Severe loss"])
        muscle_strength = random.choice(["Normal", "Slightly weaker on left side", "Slightly weaker on right side"])
        bone_structure = random.choice(["Symmetrical", "Slight asymmetry in cheekbones", "Pronounced jawline", "Rounded cheeks"])

        patient = {
            "Age": age,
            "Gender": gender,
            "Ethnicity": ethnicity,
            "Previous treatments": treatments,
            "Medical history": medical_history,
            "Allergies": allergies,
            "Skin type": skin_type,
            "Sun exposure history": sun_exposure,
            "Smoking status": smoker,
            "Primary concerns": primary_concern,
            "Secondary concerns": secondary_concern,
            "Desired outcome": outcome,
            "Budget constraints": budget,
            "Occupation": occupation,
            "Facial expressions": facial_expressions,
            "Sleep habits": sleep_habits,
            "Recent weight changes": weight_changes,
            "Hormonal status": hormonal_status,
            "Skin elasticity": skin_elasticity,
            "Facial muscle strength": muscle_strength,
            "Bone structure": bone_structure
        }
        patient_data.append(patient)

    return pd.DataFrame(patient_data)

# Generate 100 fake patient records
patients_df = generate_fake_patient_data(100)

# Write to jsonl file
file_path = "patients.jsonl" 
with open(file_path, 'w') as jsonl_file:
    for _, row in patients_df.iterrows():
        row_dict = row.to_dict()
        jsonl_file.write(f"{json.dumps(row_dict)}\n")
```
